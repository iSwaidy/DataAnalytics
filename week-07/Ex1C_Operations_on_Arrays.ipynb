```python
import numpy as np
```

## Sample Arrays

```python
r_1 = np.random.randint(1, 20, size=2)
r_2 = np.random.randint(1, 20, size=(1, 2))
r_3 = np.random.rand(2, 1)
r_4 = np.random.randint(20, 100, size=(4, 5))
r_5 = np.random.rand(4, 5)

print('r_1:', r_1)
print('r_2:', r_2)
print('r_3:', r_3)
print('r_4:', r_4)
print('r_5:', r_5)
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)
print(r_4)
```

Shuffle moves whole rows around, the individual numbers inside each row stay in the same order, but the rows themselves swap positions. It also modified r_4 directly instead of returning a new array.

```python
new_shuffle = np.random.permutation(r_4)
print(new_shuffle)
print(r_4)
```

Use shuffle when you want to scramble the original array in place and don't need to keep the old order. 

Use permutation when you want a shuffled copy but need the original array left alone.

## Mathematical Operations

```python
print('r_1:', r_1, 'shape:', r_1.shape)
print('r_2:', r_2, 'shape:', r_2.shape)
result = np.add(r_1, r_2)
print('result:', result, 'shape:', result.shape)
```

```python
print('r_1:', r_1, 'shape:', r_1.shape)
print('r_3:', r_3, 'shape:', r_3.shape)
result = np.add(r_1, r_3)
print('result:', result, 'shape:', result.shape)
```

```python
print('r_2:', r_2, 'shape:', r_2.shape)
print('r_4:', r_4, 'shape:', r_4.shape)
result = np.add(r_2, r_4)
print('result:', result, 'shape:', result.shape)
```

```python
print('r_1 shape:', r_1.shape)
print('r_4 shape:', r_4.shape)
# This errors. r_1 is (2,) and r_4 is (4,5).
# Aligning right: 2 vs 5, they're not equal and neither is 1, so broadcasting fails.
print(np.add(r_1, r_4))
```

```python
print('r_3 shape:', r_3.shape)
print('r_4 shape:', r_4.shape)
# This errors too. r_3 is (2,1) and r_4 is (4,5).
# Aligning right: 1 vs 5 (ok, one is 1), then 2 vs 4 (fail, not equal, neither is 1).
print(np.add(r_3, r_4))
```

```python
print(np.add(np.array([1]), r_4))
# Works! np.array([1]) is shape (1,). The single value 1 gets added to every element of r_4.
```

```python
print(np.add(np.array([[1]]), r_4))
# Also works. np.array([[1]]) is shape (1,1) — a 2D version of a single value.
# Same result as above, just with a different starting shape.
```

```python
print(np.add(np.array([1, 1, 1, 1, 1]), r_4))
# Works! np.array([1,1,1,1,1]) is shape (5,) — matches the 5 columns of r_4.
# Each row of r_4 gets the row [1,1,1,1,1] added to it.
```

```python
print(np.add(np.array([[1], [1]]), r_4))
# This errors. np.array([[1],[1]]) is shape (2,1).
# Aligning right: 1 vs 5, then 2 vs 4 fail.
# ValueError: operands could not be broadcast together with shapes (2,1) (4,5)
```
