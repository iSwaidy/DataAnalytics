## Looping Through Arrays

```python
arrA = np.array([[11, 12, 13], [14, 15, 16]])
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

----------------------
The first loop hits each row of the array, so it prints two arrays. 
The second loop goes one level deeper and hits each number, so it prints six values instead of two.

```python
for x in arrA:
    for y in x:
        loop_result = x * 2
print(loop_result)
# This prints the whole row doubled, not just one number.
# loop_result keeps getting overwritten with x*2, after
# the loops finish we only see the last row t
```

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
print(stacked_array_2)
# This errors out. joined_array_1 is shape (4,3) and joined_array_2 is shape (2,6),
# so they don't match. np.stack needs both arrays to have the exact same shape.
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)
```

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1)
print(another_2)
# This breaks. joined_array_1 has 4 rows, arrA has 2 rows.
# To concat side by side on axis=1, both arrays need the same number of rows.
```

```python
h_stacked = np.hstack((arrA, arrA))   # stacks horizontally
print(h_stacked)

v_stacked = np.vstack((arrA, arrA))   # stacks vertically
print(v_stacked)

d_stacked = np.dstack((arrA, arrA))   # stacks depth-wise
print(d_stacked)
```

## Splitting Arrays

```python
splitsky = np.array_split(joined_array_1, 2)
print(splitsky)
# Splits joined_array_1 into 2 arrays.
# joined_array_1 was shape (4,3), so we get two arrays of shape (2,3) each.
```

```python
spliterino = np.array_split(joined_array_1, 3)
print(spliterino)
# Splits into 3 arrays. 4 rows don't divide evenly into 3,
# so NumPy gives us two arrays with 2 rows and one with 1 row.
# Shapes: (2,3), (1,3), (1,3
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
```

## Sorting and Filtering Arrays

```python
arrB = np.array([[7, 6], [13, 5], [7, 3]])
arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
print(arrB)
z = arrB < 7
arrC = arrB[z]
print(arrC)
```

```python
print(arrB)
y = arrB >= 7
arrD = arrB[y]
print(arrD)
```
