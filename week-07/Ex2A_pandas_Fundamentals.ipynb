```python
import numpy as np
import pandas as pd
```

## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}
```

```python
print('aa:', aa)
print('ab:', ab)
print('ac:', ac)
print('ad:', ad)
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac))  # wrap the set in list() — pandas needs an ordered collection
ser4 = pd.Series(ad)
```

```python
print('ser1:')
print(ser1)
print('\nser2:')
print(ser2)
print('\nser3:')
print(ser3)
print('\nser4:')
print(ser4)
```

## DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, 'col2', ser2)
df1.insert(2, 'col3', ser3)
df1
```

```python
df1.insert(3, 'col4', ser4)
df1
# ser4's index was 'first', 'second', 'third', 'fourth'.
# df1's index is 0, 1, 2, 3.
# The indexes don't match, so pandas can't line ser4 up it fills the whole column with NaN
```

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```
