# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Styling Plots

```python
plt.plot(x, y, linestyle='--', color='red')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x, y, marker='s', markersize=10, color='green')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x, y, marker='^', color='purple')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x, y, marker='D', linestyle='--', color='orange')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# Exercise 3.B More plotting practice

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

plt.bar(products, sales)
plt.show()
```

```python
plt.bar(products, sales, color='skyblue', width=0.5)
plt.title('Product Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
sizes = np.random.rand(50) * 100
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# First subplot: sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)
axes[0].plot(x, y)
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')

# Second subplot: histogram of random numbers
data = np.random.randn(1000)
axes[1].hist(data)
axes[1].set_title('Histogram of Random Numbers')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
