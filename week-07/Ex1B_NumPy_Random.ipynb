## Generating Random Choices from an Array

```python
students = np.array(['Aspenne', 'Chantal', 'Cris', 'Dania', 'Demy', 'Dimitri',
                     'Hamzah', 'Hana', 'Ivana', 'Jonathan', 'Jordan', 'Kasaiah'])
print(students)
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4))
print(breakout_groups)
# Not a great way to assign breakout groups, np.random.choice samples WITH replacement,
# so the same person can show up multiple times while others get left out entirely.
# For real breakout groups you'd want to shuffle without replacement so everyone gets included once.
```

```python
# Probabilities have to add up to 1.0
# Aspenne = 0 (never called on)
# Cris = 0.30 (favorite, obviously)
# Everyone else splits the remaining 0.70 evenly across 10 people = 0.07 each
probs = [0.0, 0.07, 0.30, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07]
call_on_weighted = np.random.choice(students, size=5, p=probs)
print(call_on_weighted)
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```
