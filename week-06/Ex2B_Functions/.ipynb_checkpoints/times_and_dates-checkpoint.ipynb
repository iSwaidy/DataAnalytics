# Exercise 2.B Lab 2 — Working with datetime functions
This notebook explores the datetime library for working with dates and times.

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
print(today)
```

```python
print(f"Today is {today.strftime('%A, %B %d, %Y')}")
```

# My birthday

```python
my_birthday = datetime.date(1999, 1, 21)
```

```python
print(my_birthday)
```

```python
print(f"My birthday is {my_birthday.strftime('%m/%d/%Y')}")
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print(f"90 days from today is {ninety_d.strftime('%B %d, %Y')}")
```

# Time for Dinner

```python
dinner_time = datetime.time(19, 30, 0)
```

```python
print(f"Let's meet for dinner at {dinner_time.strftime('%I:%M %p')}")
```
