## Array Basics

```python
fruits = ['apple', 'banana', 'cherry', 'date']
```

```python
arr_1 = np.array(fruits)
print(arr_1)
```

```python
arr_1.dtype

# U stands for Unicode string & 6 stands for the maximum number of characters in the string
```

```python
arr_2 = np.array([(10, 11, 12), (13, 14, 15)])
print(arr_2)
```

```python
print(arr_2.T)
```

```python
print(arr_2)

# arr_2.T returns a transposed view but does NOT modify arr_2.
# The original array stays unchanged — [10 11 12] / [13 14 15]
```

```python
arr_3 = np.array([(1, 3, 5), (2, 4, 6)])
print(arr_3)
```

```python
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
```

```python
arr_5 = np.ones((2, 2, 2))
print(arr_5)
```
