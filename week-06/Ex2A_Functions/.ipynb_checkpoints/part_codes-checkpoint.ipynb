# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

```python
print('ACME:123-L'.split(':'))
print('ACME:123-L'.split('-'))
```

```python
def get_supplier_code(pc):
    return pc.split(":")[0]
```

```python
print(get_supplier_code('ACME:123-L'))
```

```python
def get_size(pc):
    return pc.split("-")[1]
```

```python
print(get_size('ACME:123-L'))
```

```python
def get_product_num(code):
    after_colon = code.split(':')[1]     
    return after_colon.split('-')[0]     
```

```python
print(get_product_num('ACME:123-L'))
```

```python
code1 = "ACME:123-L"
supplier_code1 = get_supplier_code(code1)
product_num1 = get_product_num(code1)
size1 = get_size(code1)
```

```python
print(f"For {code1}, the supplier is {supplier_code1}. Part number {product_num1} is available in size {size1}.")
```

```python
code2 = "DI:12345-M"
supplier_code2 = get_supplier_code(code2)
product_num2 = get_product_num(code2)
size2 = get_size(code2)
```

```python
print(f"For {code2}, the supplier is {supplier_code2}. Part number {product_num2} is available in size {size2}.")
```
