Creating and calling functions

```python
def favorite_things(name, movie, music):
    return f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite music is {music}."
```

```python
print(favorite_things('Cris', 'Wolf of Wall Street', 'Deftones'))
```

```python
def why_im_here(purpose):
    return f"I joined the Year Up Data Analyst program because {purpose}"
```

```python
print(why_im_here(' i want to pursue my childhood career goals in tech'))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."
```

```python
print(favorite_place('Mexico', 'I spend more time in nature with family.'))
```
